# SoccerNet UCL Downloader

Baixando apenas os jogos da UEFA Champions League nos splits `train`, `valid` e `test`.

```python
import os
from SoccerNet.Downloader import SoccerNetDownloader
from SoccerNet.utils import getListGames

UCL_CHAMPIONSHIP = "europe_uefa-champions-league"
DEFAULT_FILES = ["1_720p.mkv", "2_720p.mkv", "Labels.json"]
VALID_SPLITS = ["train", "valid", "test"]


def normalize_splits(splits):
    if not isinstance(splits, list):
        splits = [splits]

    normalized = []
    for split in splits:
        if split == "v1":
            normalized.extend(["train", "valid", "test"])
        elif split == "all":
            normalized.extend(["train", "valid", "test", "challenge"])
        else:
            normalized.append(split)
    return normalized


def get_ucl_games(splits):
    splits = normalize_splits(splits)
    games = []
    for split in splits:
        if split not in ["train", "valid", "test", "challenge"]:
            raise ValueError("Split inválido: use train, valid, test ou challenge.")

        for game in getListGames(split, task="spotting"):
            if game.startswith(UCL_CHAMPIONSHIP + os.sep) or game.startswith(UCL_CHAMPIONSHIP + "/"):
                games.append((split, game))
    return games


def download_ucl_games(local_dir, password, splits, files):
    downloader = SoccerNetDownloader(LocalDirectory=local_dir)
    if password is not None:
        downloader.password = password

    games = get_ucl_games(splits)
    if not games:
        raise RuntimeError("Nenhum jogo UEFA UCL encontrado para os splits informados.")

    print(f"Baixando {len(games)} jogos UEFA UCL nos splits: {', '.join(sorted(set(s for s, _ in games)))}")

    for split, game in games:
        print(f"-> {split}: {game}")
        downloader.downloadGame(game=game, files=files, spl=split, verbose=True)

    print("Download concluído.")
```

```python
EPL_CHAMPIONSHIP = "england_epl"  # SoccerNet's identifier for the Premier League

def get_epl_games(splits):
    splits = normalize_splits(splits)
    games = []
    for split in splits:
        if split not in ["train", "valid", "test", "challenge"]:
            raise ValueError("Split inválido: use train, valid, test ou challenge.")
        for game in getListGames(split, task="spotting"):
            if game.startswith(EPL_CHAMPIONSHIP + os.sep) or game.startswith(EPL_CHAMPIONSHIP + "/"):
                games.append((split, game))
    return games


def count_epl_games(splits):
    games = get_epl_games(splits)
    split_counts = {}
    for split, _ in games:
        split_counts[split] = split_counts.get(split, 0) + 1

    print(f"Total de jogos EPL: {len(games)}")
    for split, count in sorted(split_counts.items()):
        print(f"  {split}: {count} jogos")

    return games
```
