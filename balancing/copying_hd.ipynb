```python
import pandas as pd
import shutil
import os
import glob

USED_CSVS_GLOB = "/mnt/storage_C4/gaussian_football/data/balanced_datasets/used/*.csv"
HD_ROOT = "/media/al.pedro.santos/Seagate/gaussian_football/data/processed"
DEST_ROOT = "/mnt/storage_C4/gaussian_football/data/processed"
DRY_RUN = True

MARKER = "data/processed/"

def relative_suffix(path):
    idx = path.find(MARKER)
    return path[idx + len(MARKER):] if idx != -1 else path.lstrip("/")

used_paths = set()
for csv_path in glob.glob(USED_CSVS_GLOB):
    df = pd.read_csv(csv_path)
    for col in ("clip_path", "mel_path"):
        valid = df[col].dropna()
        valid = valid[valid.str.len() > 0]
        used_paths.update(valid.tolist())

print(f"Total de arquivos a copiar (video + mel): {len(used_paths)}")

copied, missing = 0, []
for path in used_paths:
    rel = relative_suffix(path)
    src = os.path.join(HD_ROOT, rel)
    dst = os.path.join(DEST_ROOT, rel)

    if not os.path.exists(src):
        missing.append(src)
        continue

    if DRY_RUN:
        copied += 1
        continue

    os.makedirs(os.path.dirname(dst), exist_ok=True)
    shutil.copy2(src, dst)
    copied += 1

print(f"{'Seriam copiados' if DRY_RUN else 'Copiados'}: {copied}")
print(f"Faltando no HD: {len(missing)}")
if missing:
    with open("missing_clips.txt", "w") as f:
        f.write("\n".join(missing))
```

```python
MARKER = "data/processed/"

def relative_suffix(path):
    idx = path.find(MARKER)
    return path[idx + len(MARKER):] if idx != -1 else path.lstrip("/")

with open("files_to_copy.txt", "w") as f:
    for path in used_paths:
        f.write(relative_suffix(path) + "\n")

print(f"{len(used_paths)} paths escritos em files_to_copy.txt")
```
