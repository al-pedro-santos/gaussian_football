## AVALIAÇÃO

#### Calcular o CCC

Para validar se a função gerada pela rede é boa, a forma ideal de comparar com o seu Ground Truth (que também deve ser mapeado para o tempo) é usando o CCC, que mede a concordância de trajetórias temporais:

$$ρ_c​= \dfrac{2ρσ_x​σ_y}{σ_x^2​+σ_y^2+(μ_x​−μ_y​)^2}​​$$

Onde ρ é a correlação de Pearson, μ é a média e σ2 é a variância. O CCC pune o modelo não apenas se a curva subir e descer nos momentos errados, mas também se houver um deslocamento de escala (ex: o modelo captou a variação, mas os valores absolutos estão muito baixos).

```python
def calculate_ccc(y_true, y_pred):
    """Calcula o Concordance Correlation Coefficient (CCC)."""
    mean_true = np.mean(y_true)
    mean_pred = np.mean(y_pred)
    var_true = np.var(y_true)
    var_pred = np.var(y_pred)
    
    # Covariância
    covariance = np.mean((y_true - mean_true) * (y_pred - mean_pred))
    
    # Fórmula do CCC
    numerator = 2 * covariance
    denominator = var_true + var_pred + (mean_true - mean_pred) ** 2
    
    if denominator == 0:
        return 0.0
    return numerator / denominator
```

#### Vídeo para Arousal

```python
import torch
import numpy as np

from scipy.interpolate import interp1d
from scipy.stats import pearsonr

def evaluate_video_to_arousal_fn(model, video_frames, fps, true_arousal_values, plot=False, device=device):
    """
    Avalia o modelo em um vídeo e retorna uma função contínua de arousal f(t).
    
    Argumentos:
        model: O modelo de Deep Learning (ex: 3D-CNN, RNN, Transformer).
        video_frames: Tensor do vídeo com formato (1, Tempo, Canais, Altura, Largura).
        fps: Frames por segundo do vídeo original (int ou float).
        true_arousal_values: Array/Lista com os valores de arousal verdadeiro correspondentes.
        plot: booleano que decide se vamos printar nossos resultados ou não
    
    Retorna:
        arousal_fn: Uma função que aceita o tempo t (em segundos) e retorna o arousal.
    """
    model.eval()
    model.to(device)
    
    #colocar o vídeo no device
    video_input = video_frames        
    video_input = video_input.to(device)
    
    with torch.no_grad():
        
        # A rede deve retornar um score de arousal para cada passo temporal
        # Output esperado: (1, num_timestamps) ou (num_timestamps,)

        #Aqui vai depender do output da nossa rede, tem que ver oq o piguel fez.
        outputs = model(video_input)
        outputs = outputs.squeeze().cpu().numpy()
    
    num_timestamps = len(outputs)
    #descobre o tamanho do vídeo
    total_duration = num_timestamps / fps
    time_steps = np.linspace(0, total_duration, num_timestamps)
    
    # Cria a função contínua f(t) usando interpolação linear
    # bounds_error=False preenche com o valor mais próximo caso t saia do limite do vídeo
    pred_fn = interp1d(time_steps, outputs, kind='linear', bounds_error=False, fill_value="extrapolate")
    
    y_true = np.array(true_arousal_values)
    y_pred = pred_fn(num_timestamps)
    
    # 3. Cálculo das Métricas
    mse = np.mean((y_true - y_pred) ** 2)
    mae = np.mean(np.abs(y_true - y_pred))
    pearson_corr, _ = pearsonr(y_true, y_pred)
    ccc = calculate_ccc(y_true, y_pred)
    
    metrics = {
        "MSE": mse,
        "MAE": mae,
        "Pearson_R": pearson_corr,
        "CCC": ccc
    }
    
    if plot:
        plt.figure(figsize=(10, 4))
        plt.plot(num_timestamps, y_true, label="Verdadeiro", color="black")
        plt.plot(num_timestamps, y_pred, label="Predito", color="crimson", linestyle="--")
        plt.title(f"CCC: {ccc:.4f} | MSE: {mse:.4f}")
        plt.legend()
        plt.show()

    return metrics
```

## TREINAMENTO

```python
import os
import datetime
import torch
from torch.utils.tensorboard import SummaryWriter

def train_model(model, 
                train_loader, 
                val_samples, 
                optimizer, 
                criterion, 
                epochs=10, 
                device="cuda", 
                patience=5):
    """
    Treina a rede e a avalia a cada época usando a função contínua de arousal.
    
    Argumentos:
        model: O modelo PyTorch.
        train_loader: DataLoader do PyTorch para os dados de treino.
        val_samples: Lista de dicionários contendo os dados de validação de cada vídeo:
                     [{"video": tensor, "fps": int, "times": list, "values": list}, ...]
        optimizer: Otimizador (ex: Adam).
        criterion: Função de perda do treino (ex: nn.MSELoss()).
        patience: Número de épocas para esperar sem melhora no CCC antes de parar o treino.
    """
    model.to(device)
    best_val_ccc = -1.0  # O CCC varia de -1 a 1 (queremos o mais próximo de 1)
    
    # Configuração do Early Stopping
    epochs_no_improve = 0
    early_stop = False

    #usar o TensorBoard
    current_time = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
    log_dir = os.path.join("runs", f"arousal_experiment_{current_time}")
    writer = SummaryWriter(log_dir=log_dir)

    print(f"--> TensorBoard inicializado! Logs sendo salvos em: {log_dir}")

    for epoch in range(epochs):
        # --- FASE DE TREINO ---
        model.train()
        running_loss = 0.0
        
        for batch_videos, batch_targets in train_loader:
            batch_videos = batch_videos.to(device)
            batch_targets = batch_targets.to(device)
            
            optimizer.zero_grad()
            outputs = model(batch_videos).squeeze()
            
            loss = criterion(outputs, batch_targets)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item() * batch_videos.size(0)
            
        epoch_train_loss = running_loss / len(train_loader.dataset)
        
        #Envia a perda de treino para o TensorBoard
        writer.add_scalar("Loss/Train", epoch_train_loss, epoch)

        # --- FASE DE VALIDAÇÃO (Usando o Evaluate customizado) ---
        val_metrics_accum = {"MSE": 0, "MAE": 0, "Pearson_R": 0, "CCC": 0}
        num_val_videos = len(val_samples)
        
        for sample in val_samples:
            # Executa o evaluate SEM plotar o gráfico
            metrics = evaluate_video_to_arousal_fn(
                model=model,
                video_frames=sample["video"],
                fps=sample["fps"],
                true_times=sample["times"],
                true_values=sample["values"],
                device=device,
                plot=False #sem gráficos
            )
            # Acumula as métricas de cada vídeo
            for key in val_metrics_accum:
                val_metrics_accum[key] += metrics[key]
        
        # Tira a média das métricas entre todos os vídeos de validação
        avg_metrics = {k: v / num_val_videos for k, v in val_metrics_accum.items()}
        
        # Envia as métricas para o TensorBoard
        writer.add_scalar("Loss/Validation_MSE", avg_metrics["MSE"], epoch)
        writer.add_scalar("Metrics/MAE", avg_metrics["MAE"], epoch)
        writer.add_scalar("Metrics/Pearson", avg_metrics["Pearson_R"], epoch)
        writer.add_scalar("Metrics/CCC", avg_metrics["CCC"], epoch)

        print(f"Época [{epoch+1}/{epochs}] "
              f"| Train Loss: {epoch_train_loss:.4f} "
              f"| Val MSE: {avg_metrics['MSE']:.4f} "
              f"| Val Pearson: {avg_metrics['Pearson_R']:.4f} "
              f"| Val CCC: {avg_metrics['CCC']:.4f}")
        
        # --- SALVAR O MELHOR MODELO BASEADO NO CCC ---
        if avg_metrics["CCC"] > best_val_ccc:
            best_val_ccc = avg_metrics["CCC"]
            torch.save(model.state_dict(), "best_arousal_model.pth")
            print(f"--> Novo melhor modelo salvo! (Val CCC: {best_val_ccc:.4f})")
            epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            print(f"--> Sem melhora no CCC por {epochs_no_improve}/{patience} épocas.")
            if epochs_no_improve >= patience:
                print(f" Early Stopping engatilhado! O modelo parou de evoluir.")
                early_stop = True
                
        if early_stop:
            break
    
    writer.close()
    print("\nTreinamento concluído!")
```
